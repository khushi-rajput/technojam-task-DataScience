```python
#import packages for data analysis
import pandas as pd # pandas is used for Data Manipulation and Analysis
import matplotlib.pyplot as plt # matplotlib is used for visualizing the data in the graphical form
import plotly.express as px # exploratory data analysis(EDA)
```

```python
#read the dataset into a Notebook
df = pd.read_csv("Brasile-real-estate-dataset.csv", encoding='cp1252')
df.head() # by default it returns the first five rows
df.tail() # by default it returns the last five rows
```

```python
#Clean the data if necessary (handle missing values, outliers, etc.)
df.dropna(subset = ['area_m2'], inplace = True) # dropna used for removes the rows that contain nan values
df
```

```python
# shape of the data
df.shape # it returns the total rows and columns in our data
```

```python
# it displays the information of our data
df.info()
```

```python
# We wll see a map that's centered on Brazil City, and we can use the "Zoom Out" button in the upper-right corner of the map so that we can see the whole country.
fig = px.scatter_mapbox(
    df, # our dataset
    lat = 'lat',
    lon = 'lon',
    center = {"lat": 19.43, "lon": -99.13}, # map will be centered on brazil city
    width = 600, # width of map
    height = 600, # height of map
    hover_data = ["price_brl"], # display price when hovering mouse over house
)

fig.update_layout(mapbox_style = "open-street-map")
fig.show()
```

```python
# Perform basic data visualization to gain insights.

# Histogram of "area_m2"

#plt.hist(df['area_m2']);
plt.hist(df['area_m2']) # semicolon is used for hide the unwanted data
plt.xlabel("Area [sq meters]")
plt.ylabel("Frequency")
plt.title("Distribution of Home Sizes")
```

```python
# Horizontal boxplot of "area_m2"

plt.boxplot(df['area_m2'], vert = False)
plt.xlabel("Area [sq meters]")
plt.ylabel("Frequency")
plt.title("Distribution of Home Sizes")
```
